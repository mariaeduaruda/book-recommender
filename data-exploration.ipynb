```python
import kagglehub

# Set the path to the file you'd like to load
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Caminho para o dataset", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

# Define os rótulos dos eixos para o gráfico.
plt.xlabel("Linhas do Dataset")
plt.ylabel("Colunas com Dados Faltantes")
```

```python
import numpy as np
# A nova coluna chamada "missing_description", terá o valor 1 se a descrição do livro estiver faltando (isna) e 0 caso contrário.
# Isso ajuda a analisar se a falta de descrição tem algum impacto.
books ["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
# Define uma lista com as colunas que queremos analisar a correlação.
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# Calcula a matriz de correlação de Spearman entre as colunas de interesse.
correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Correlação de Spearman"})
heatmap.set_title("Heatmap de Correlação")
plt.show()
```

```python
# Um novo DataFrame chamado `book_missing` vai conter apenas as linhas (livros) que NÃO têm valores faltando.
# O símbolo `~` inverte a condição, ou seja, seleciona o que NÃO é nulo.
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
# Uma nova coluna chamada "words_in_description" divide a descrição em uma lista de palavras e conta o número de palavras nessa lista.
# Isso ajuda a entender o tamanho das descrições.
book_missing["words_in_description"] =(
    book_missing["description"].str.split().str.len())
```

```python
# Um novo DataFrame final, que contem apenas os livros que têm uma descrição com 25 ou mais palavras. Para garantir apenas descrições de boa qualidade.
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
# A nova coluna "title_and_subtitle" recebe o título e o subtítulo caso esse não exista, a coluna receberá o título.

book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
